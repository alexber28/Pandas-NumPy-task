**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.array([
    [1, 2, 3],
    [1, 1, 1],
    [4, 5, 6],
    [2, 4, 9], 
    [0, 0, 0],
    [1, 5, 9],
    [7, 8, 9],
    [3, 3, 3]
])

B = np.array([
    [1, 2],
    [4, 5]
])

mask1 = np.isin(A, B[0]).any(axis=1) 
mask2 = np.isin(A, B[1]).any(axis=1)

final_mask = mask1 & mask2

result = A[final_mask]
print("Подходящие строки из A:\n", result)
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
A = np.random.randint(0, 2, (10, 3))

mask = ~np.all(A == A[:, [0]], axis=1)

result = A[mask]

print("Исходная матрица:\n", A)
print("\nРезультат (только разные значения):\n", result)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
x = [2, 2, 2, 3, 3, 3, 5]
nums = []
length = []
i = 0
while i <= len(x)-1:
    nums.append(x[i])
    counter = 1
    j = i + 1
    while j <= len(x)-1 and x[j] == x[i]:
        counter += 1
        j += 1
    length.append(counter)
    i += counter
print(f"nums: {nums} \nlength: {length}")
```

```python
def rle_numpy(x):
    locs = np.where(np.diff(x) != 0)[0]
    pos = np.concatenate([[-1], locs, [len(x) - 1]])
    values = x[pos[1:]]
    counts = np.diff(pos)
    
    return values, counts

x_np = np.array([2, 2, 2, 3, 3, 3, 5])
vals, cnts = rle_numpy(x_np)
print((vals, cnts))
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
import math
def euclidean_python(X, Y):
    M, K = len(X), len(X[0])
    N = len(Y)
    dist_matrix = [[0.0 for _ in range(N)] for _ in range(M)]
    
    for i in range(M):
        for j in range(N):
            squared_sum = 0
            for k in range(K):
                squared_sum += (X[i][k] - Y[j][k])**2
            dist_matrix[i][j] = math.sqrt(squared_sum)
    return dist_matrix

# Выборка X: 3 объекта, у каждого по 2 координаты (X, Y)
X_py = [
    [0, 0],   # Точка A
    [1, 1],   # Точка B
    [2, 2]    # Точка C
]

# Выборка Y: 2 объекта, с которыми мы хотим сравнить X
Y_py = [
    [1, 0],   # Точка D
    [0, 1]    # Точка E
]

# Результатом будет матрица 3x2 (расстояния A-D, A-E, B-D, B-E...)
result = euclidean_python(X_py, Y_py)
result
```

```python
def euclidean_numpy(X, Y):
    # 1. Сумма квадратов по строкам для X и Y
    # (X**2).sum(axis=1) дает вектор (M,)
    # reshape(-1, 1) делает его столбцом (M, 1) для broadcasting
    x_2 = np.sum(X**2, axis=1).reshape(-1, 1)
    y_2 = np.sum(Y**2, axis=1)
    
    # 2. Матричное произведение X * Y.T (размер M x N)
    xy = X @ Y.T
    
    # 3. Собираем формулу: x^2 - 2xy + y^2
    # Благодаря broadcasting x_2 (столбец) и y_2 (строка) растянутся в матрицу
    dists_sq = x_2 - 2*xy + y_2
    
    # Из-за погрешности float могут появиться крошечные отрицательные числа
    # Ограничим их снизу нулем перед взятием корня
    return np.sqrt(np.maximum(dists_sq, 0))

x_arr = np.array([
    [0, 0],   # Точка A
    [1, 1],   # Точка B
    [2, 2]    # Точка C
])
y_arr = np.array([
    [1, 0],   # Точка D
    [0, 1]    # Точка E
])
print(euclidean_numpy(x_arr, y_arr))
```
